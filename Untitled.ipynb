# 🔧 Feature Preprocessing for Model Building

Now that we have cleaned data, we need to **preprocess** the features for our recommendation models:
- Parse JSON-like string columns (genres, keywords, cast, directors) into actual Python lists
- Select and prepare only the features needed for modeling
- Create the final preprocessed dataset for the recommendation engine

```python
# Load the cleaned dataset
df_preprocessed = pd.read_csv("movies_cleaned_full.csv", low_memory=False)

print("=" * 60)
print("📂 LOADED CLEANED DATASET FOR PREPROCESSING")
print("=" * 60)
print(f"Total records: {len(df_preprocessed):,}")
print(f"Total features: {df_preprocessed.shape[1]}")
print("=" * 60)
```

```python
# Define parsing function for JSON-like string columns
def parse_list_column(x):
    """
    Parse JSON-like string representations into Python lists
    Handles formats like: "['Action', 'Drama']" or "Action, Drama"
    """
    if isinstance(x, list):
        return x
    if pd.isna(x) or x == "":
        return []
    if isinstance(x, str):
        # Try to parse as literal (JSON-like format)
        try:
            parsed = ast.literal_eval(x)
            if isinstance(parsed, list):
                return [str(item).strip() for item in parsed]
        except:
            pass
        # Fallback: split by comma
        return [item.strip() for item in str(x).split(",")]
    return []

print("✅ Parsing function defined")
print("=" * 60)
```

```python
# Parse list-type columns (genres, keywords, cast, directors, etc.)
print("🔄 Parsing JSON-like string columns into Python lists...")
print("=" * 60)

list_columns = ['genres', 'keywords', 'cast', 'directors', 'writers', 
                'production_companies', 'production_countries', 'spoken_languages']

parsed_count = 0
for col in list_columns:
    if col in df_preprocessed.columns:
        print(f"   Processing: {col}...")
        df_preprocessed[col] = df_preprocessed[col].apply(parse_list_column)
        parsed_count += 1
        
        # Show sample
        sample = df_preprocessed[col].iloc[0]
        print(f"      Sample: {sample[:3] if len(sample) > 0 else 'Empty'}")

print(f"\n✅ Parsed {parsed_count} columns successfully")
print("=" * 60)
```

```python
# Verify parsing worked correctly
print("🔍 Verification: Checking parsed data types...")
print("=" * 60)

for col in list_columns:
    if col in df_preprocessed.columns:
        # Check if all values are lists
        is_list = df_preprocessed[col].apply(lambda x: isinstance(x, list)).all()
        avg_length = df_preprocessed[col].apply(len).mean()
        print(f"   {col}: {'✅ List' if is_list else '❌ Not List'} | Avg length: {avg_length:.1f}")

print("=" * 60)
```

```python
# Select features for the recommendation model
print("Selecting features for model dataset...")
print("=" * 60)

model_features = [
    "title", 
    "genres", 
    "keywords", 
    "cast", 
    "directors", 
    "overview", 
    "vote_average", 
    "runtime", 
    "release_date", 
    "poster_path"
]

# Keep only available features
available_features = [col for col in model_features if col in df_preprocessed.columns]
df_model = df_preprocessed[available_features].copy()

print(f"Selected {len(available_features)} features:")
for feat in available_features:
    print(f"   ✓ {feat}")

print(f"\n📊 Model dataset shape: {df_model.shape}")
print("=" * 60)
```

```python
# Final quality filtering for model dataset
print("🎯 Final Quality Filtering...")
print("=" * 60)

initial_count = len(df_model)

# Must have title
df_model = df_model[df_model["title"].notna() & (df_model["title"] != "")]
print(f"   ✅ Movies with titles: {len(df_model):,}")

# Must have meaningful overview (description)
df_model["overview"] = df_model["overview"].fillna("")
df_model = df_model[df_model["overview"].str.len() > 10]
print(f"   ✅ Movies with descriptions: {len(df_model):,}")

# Filter by rating (keep only 6.0+)
if 'vote_average' in df_model.columns:
    df_model = df_model[df_model['vote_average'] >= 6.0]
    print(f"   ✅ Movies with rating >= 6.0: {len(df_model):,}")

df_model = df_model.reset_index(drop=True)

print(f"\n📉 Removed {initial_count - len(df_model):,} low-quality entries")
print(f"📊 Final model dataset: {len(df_model):,} movies")
print("=" * 60)
```

# 🤖 Model Building & Feature Engineering

Now we'll build two recommendation systems:
1. **Content-Based Filtering** (TF-IDF + Cosine Similarity)
2. **Simple KNN** (Binary Features + Cosine Distance)

```python
# Content-Based Filtering System (TF-IDF + Cosine Similarity)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("🔧 Building Content-Based Recommendation System (TF-IDF)")
print("=" * 60)

# Helper function to clean strings (remove spaces for better matching)
def clean_string(text):
    return str(text).replace(' ', '').replace('-', '')

# Create "soup" - weighted combination of features
def create_soup(row):
    """
    Combine movie features with optimal weights:
    - Genres: 7x weight
    - Directors: 7x weight  
    - Cast: 2x weight (top 5 actors)
    - Keywords: 3x weight (top 10 keywords)
    - Overview: 1x weight
    """
    genres = [clean_string(g) for g in row['genres']]
    directors = [clean_string(d) for d in row['directors']]
    cast = [clean_string(c) for c in row['cast'][:5]]
    keywords = [clean_string(k) for k in row['keywords'][:10]]
    overview_words = clean_string(row['overview']).split()
    
    parts = []
    parts.extend(genres * 7)
    parts.extend(directors * 7)
    parts.extend(cast * 2)
    parts.extend(keywords * 3)
    parts.extend(overview_words * 1)
    
    return ' '.join(parts)

# Apply to all movies
df_model['soup'] = df_model.apply(create_soup, axis=1)

# Remove movies with empty features
df_model = df_model[df_model['soup'].str.strip() != ''].reset_index(drop=True)
print(f"📊 Movies with features: {len(df_model):,}")

# Build TF-IDF matrix
print("🔄 Building TF-IDF matrix...")
tfidf_vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=50000,
    ngram_range=(1, 2),  # Use unigrams and bigrams
    min_df=2             # Minimum document frequency
)

tfidf_matrix = tfidf_vectorizer.fit_transform(df_model['soup'])

# Create title to index mapping for fast lookups
title_to_idx = pd.Series(df_model.index, index=df_model['title'].str.lower()).to_dict()

print(f"✅ TF-IDF matrix shape: {tfidf_matrix.shape}")
print(f"✅ Vocabulary size: {len(tfidf_vectorizer.vocabulary_):,}")
print(f"✅ Movies indexed: {len(title_to_idx):,}")
print("=" * 60)
```

```python
# Simple KNN-Based Recommendation System 
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MultiLabelBinarizer, normalize
from scipy import sparse

print("🔧 Setting up Simple KNN-Based Recommendation System...")
print("=" * 100)

# Initialize binarizers for each feature type
mlb_genres = MultiLabelBinarizer(sparse_output=True)
mlb_directors = MultiLabelBinarizer(sparse_output=True)
mlb_cast = MultiLabelBinarizer(sparse_output=True)
mlb_keywords = MultiLabelBinarizer(sparse_output=True)

# Prepare data with more features for better recommendations
genres_data = df_model['genres'].tolist()
directors_data = [directors[:3] if isinstance(directors, list) else [] for directors in df_model['directors']]
cast_data = [cast[:5] if isinstance(cast, list) else [] for cast in df_model['cast']]
keywords_data = [keywords[:10] if isinstance(keywords, list) else [] for keywords in df_model['keywords']]

print("📊 Encoding features using MultiLabelBinarizer (sparse matrices)...")

# Fit and transform (one-hot encode) - returns sparse matrices
genres_encoded = mlb_genres.fit_transform(genres_data)
directors_encoded = mlb_directors.fit_transform(directors_data)
cast_encoded = mlb_cast.fit_transform(cast_data)
keywords_encoded = mlb_keywords.fit_transform(keywords_data)

print(f"✅ Genres: {genres_encoded.shape[1]} unique values")
print(f"✅ Directors: {directors_encoded.shape[1]} unique values (top 3 per movie)")
print(f"✅ Cast: {cast_encoded.shape[1]} unique values (top 5 per movie)")
print(f"✅ Keywords: {keywords_encoded.shape[1]} unique values (top 10 per movie)")

# Combine all features horizontally (sparse matrices) with weights
feature_matrix_knn = sparse.hstack([
    genres_encoded * 3,      # Weight genres more
    directors_encoded * 3,   # Weight directors more
    cast_encoded,            # Cast weight = 1
    keywords_encoded * 2     # Weight keywords moderately
])

# NORMALIZE the feature vectors (important for cosine distance!)
feature_matrix_knn = normalize(feature_matrix_knn, norm='l2', axis=1)

print(f"\n✅ Feature matrix shape: {feature_matrix_knn.shape}")
print(f"   Format: Sparse CSR matrix (L2 normalized)")
print(f"   Memory: ~{feature_matrix_knn.data.nbytes / 1024**2:.1f} MB")

# Train KNN model with COSINE distance (works better with normalized binary features)
knn_simple = NearestNeighbors(
    n_neighbors=11,
    metric='cosine',  # Cosine distance (1 - cosine similarity)
    algorithm='brute',
    n_jobs=-1
)

print("🔄 Training KNN model with Cosine distance...")
knn_simple.fit(feature_matrix_knn)

print(f"✅ Simple KNN Model trained on {feature_matrix_knn.shape[0]:,} movies")
print(f"✅ Using COSINE distance on normalized weighted binary features")
print(f"   (Genres×3, Directors×3, Cast×1, Keywords×2)")
print("=" * 100)
```

```python
# COMPARISON: Content-Based (TF-IDF) vs Simple KNN (Binary Features)
print("⚖️ ALGORITHM COMPARISON - CONTENT-BASED (TF-IDF) vs SIMPLE KNN (NO TF-IDF)")
print("=" * 100)

test_movies = ["Spider-Man", "The Avengers"]  # CORRECTED!

print(f"\n🎬 Test Movies: {', '.join(test_movies)}")
print("=" * 100)

try:
    # Find indices
    indices = []
    for movie in test_movies:
        title_lower = movie.lower()
        if title_lower in title_to_idx:
            indices.append(title_to_idx[title_lower])
    
    if len(indices) >= 2:
        print("\n" + "="*100)
        print("📊 CONTENT-BASED FILTERING (TF-IDF + Cosine Similarity)")
        print("   • Uses: Text analysis of plot, keywords, genres, directors, cast")
        print("   • Method: TF-IDF vectorization + Cosine similarity")
        print("="*100)
        
        # Content-Based
        favorite_vectors = tfidf_matrix[indices]
        avg_vector = np.asarray(favorite_vectors.mean(axis=0))
        cosine_sims = cosine_similarity(avg_vector, tfidf_matrix).flatten()
        cb_indices = cosine_sims.argsort()[::-1]
        cb_indices = [i for i in cb_indices if i not in indices][:10]
        
        cb_results = df_model.iloc[cb_indices][['title', 'vote_average', 'genres']].copy()
        cb_results['match_%'] = [round(cosine_sims[i] * 100, 1) for i in cb_indices]
        cb_results['genres'] = cb_results['genres'].apply(lambda x: ', '.join(x[:2]) if isinstance(x, list) else str(x))
        
        print(cb_results.to_string(index=False))
        
        print("\n" + "="*100)
        print("📊 SIMPLE KNN (Binary Features + Cosine Distance - NO TF-IDF)")
        print("   • Uses: One-hot encoded genres, directors, cast, keywords")
        print("   • Method: K-Nearest Neighbors with Cosine distance on normalized weighted binary features")
        print("="*100)
        
        # Simple KNN - Get feature vectors for input movies
        favorite_features = feature_matrix_knn[indices]
        # Average the features (convert to array to avoid matrix issues)
        avg_features = np.asarray(favorite_features.mean(axis=0)).flatten()
        # Normalize - reshape to 2D for normalize function
        avg_features = normalize(avg_features.reshape(1, -1), norm='l2')
        
        # Find neighbors (cosine distance)
        distances, knn_indices = knn_simple.kneighbors(avg_features, n_neighbors=12)
        knn_indices = knn_indices.flatten()
        knn_filtered = [idx for idx in knn_indices if idx not in indices][:10]
        
        knn_results = df_model.iloc[knn_filtered][['title', 'vote_average', 'genres']].copy()
        
        # Convert cosine distance to similarity percentage
        # cosine distance = 1 - cosine similarity, so similarity = 1 - distance
        filtered_dists = [distances.flatten()[list(knn_indices).index(idx)] for idx in knn_filtered]
        knn_results['match_%'] = [round((1 - dist) * 100, 1) for dist in filtered_dists]
        knn_results['genres'] = knn_results['genres'].apply(lambda x: ', '.join(x[:2]) if isinstance(x, list) else str(x))
        
        print(knn_results.to_string(index=False))
        
        # Calculate metrics
        print("\n" + "="*100)
        print("📈 QUALITY METRICS")
        print("="*100)
        
        cb_avg_rating = cb_results['vote_average'].mean()
        knn_avg_rating = knn_results['vote_average'].mean()
        cb_avg_match = cb_results['match_%'].mean()
        knn_avg_match = knn_results['match_%'].mean()
        
        print(f"\nContent-Based (TF-IDF + Cosine):")
        print(f"  • Average Rating: {cb_avg_rating:.2f}")
        print(f"  • Average Match: {cb_avg_match:.1f}%")
        
        print(f"\nSimple KNN (Binary Features + Cosine):")
        print(f"  • Average Rating: {knn_avg_rating:.2f}")
        print(f"  • Average Match: {knn_avg_match:.1f}%")
        
        print(f"\n🏆 WINNER:")
        if cb_avg_rating > knn_avg_rating and cb_avg_match > knn_avg_match:
            print("   Content-Based (Better ratings AND match scores)")
        elif knn_avg_rating > cb_avg_rating and knn_avg_match > cb_avg_match:
            print("   Simple KNN (Better ratings AND match scores)")
        elif cb_avg_rating > knn_avg_rating:
            print("   Content-Based (Higher average rating)")
        elif knn_avg_rating > cb_avg_rating:
            print("   Simple KNN (Higher average rating)")
        else:
            print("   TIE (Similar performance)")
        
        # Show key differences
        print(f"\n📋 KEY DIFFERENCES:")
        print(f"   Content-Based (TF-IDF):")
        print(f"      - Uses TF-IDF to understand IMPORTANCE of words in plot descriptions")
        print(f"      - Captures semantic similarity through text analysis")
        print(f"      - Better for finding thematically similar movies")
        print(f"      - Includes plot/overview text in analysis")
        print(f"\n   Simple KNN (Binary Features):")
        print(f"      - Uses simple one-hot encoding (movie has feature or not)")
        print(f"      - NO text analysis - just direct feature matching")
        print(f"      - Cosine distance on normalized weighted binary features")
        print(f"      - Feature weights: Genres×3, Directors×3, Cast×1, Keywords×2")
        print(f"      - More interpretable: matches based on concrete attributes")
        print(f"      - Ignores plot descriptions completely!")
        
except Exception as e:
    print(f"\n❌ Error: {e}")
    import traceback
    traceback.print_exc()

print("\n" + "=" * 100)
print("✅ Comparison complete!")
```

```python
# SIMPLE COMPARISON: Which Algorithm Gives More Relevant Recommendations?
print("📊 SIMPLE PROOF: Content-Based vs KNN")
print("=" * 60)

test_movies = ["Fight Club", "Kill Bill: Vol. 1"]
indices = [title_to_idx[m.lower()] for m in test_movies]

# Get 20 recommendations from each
# Content-Based
fav_vec = np.asarray(tfidf_matrix[indices].mean(axis=0))
cosine_sims = cosine_similarity(fav_vec, tfidf_matrix).flatten()
cb_indices = cosine_sims.argsort()[::-1]
cb_indices = [i for i in cb_indices if i not in indices][:12]
cb_scores = [cosine_sims[i] * 100 for i in cb_indices]

# KNN
fav_feat = np.asarray(feature_matrix_knn[indices].mean(axis=0)).flatten()
fav_feat = normalize(fav_feat.reshape(1, -1), norm='l2')
distances, knn_idx = knn_simple.kneighbors(fav_feat, n_neighbors=22)
knn_indices = [idx for idx in knn_idx.flatten() if idx not in indices][:12]
knn_scores = [(1 - distances.flatten()[list(knn_idx.flatten()).index(idx)]) * 100 for idx in knn_indices]

# Simple bar chart
fig, ax = plt.subplots(figsize=(12, 6))

x = range(1, 13)
width = 0.35

bars1 = ax.bar([i - width/2 for i in x], cb_scores, width, label='Content-Based', color='#27ae60', alpha=0.8)
bars2 = ax.bar([i + width/2 for i in x], knn_scores, width, label='KNN', color='#95a5a6', alpha=0.8)

# Add average lines
cb_avg = np.mean(cb_scores)
knn_avg = np.mean(knn_scores)
ax.axhline(cb_avg, color='#27ae60', linestyle='--', linewidth=2, label=f'CB Average: {cb_avg:.1f}%')
ax.axhline(knn_avg, color='#95a5a6', linestyle='--', linewidth=2, label=f'KNN Average: {knn_avg:.1f}%')

ax.set_xlabel('Recommendation Rank', fontsize=12, fontweight='bold')
ax.set_ylabel('Similarity Score (%)', fontsize=12, fontweight='bold')
ax.set_title('Which Algorithm Gives More Relevant Recommendations?\n(Higher = Better Match to Your Taste)', 
             fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)

# Add winner text
improvement = ((cb_avg - knn_avg) / knn_avg) * 100
winner_text = f'🏆 Content-Based is {improvement:.1f}% MORE RELEVANT!'
ax.text(0.5, 0.95, winner_text, transform=ax.transAxes, fontsize=13, fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.8', facecolor='#d5f4e6', edgecolor='#27ae60', linewidth=3),
        ha='center', va='top')

plt.tight_layout()
plt.show()

print(f"\n✅ RESULTS:")
print(f"   Content-Based Average: {cb_avg:.1f}%")
print(f"   KNN Average: {knn_avg:.1f}%")
print(f"   🏆 Content-Based is {improvement:.1f}% better!")
print("=" * 60)
```
